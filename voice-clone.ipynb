# Importing modules

```python
import numpy as np 
import matplotlib.pyplot as plt 
import os 
import librosa  
import librosa.display
import IPython.display as ipd 
```

```python
BASE_DIR = "data/archive/data/"

TRAIN_DIR = "TRAIN/"
```

```python
FIRST_TEST_FILE = "DR1/FCJF0/SA1.WAV.wav"
```

```python
first_test = BASE_DIR + TRAIN_DIR + FIRST_TEST_FILE
```

```python
ipd.Audio(first_test)
```

```python
test_fft, sr = librosa.load(first_test)
```

```python
FRAME_SIZE = 2048
HOP_SIZE = 256
```

```python
S_scale = librosa.stft(test_fft, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)
```

```python
S_scale.shape
```

```python
Y_scale = np.abs(S_scale) ** 2
```

```python
Y_log_scale = librosa.power_to_db(Y_scale)
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(Y_log_scale, sr=sr, hop_length=HOP_SIZE, x_axis="time", y_axis="log")
plt.colorbar(format="%+2.f")
```

```python
mel_spectrogram = librosa.feature.melspectrogram(y=test_fft, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_SIZE, n_mels=90)
```

```python
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(log_mel_spectrogram, x_axis="time", y_axis="mel", sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```
